# Исследование объявлений о продаже квартир

На основании данных сервиса Яндекс.Недвижимость (архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет) нужно проанализировать влияние различных факторов на ценообразование в СпБ. Нужно научить систему прогносзировать рыночную стоимость объектов недвижимости. Установить ключевые параметры которые определяют стоимость квартиры параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 


## Изучение данных из файла

```python
import pandas as pd
#выводим округленные данные и отменим экспоненциальную запись
pd.set_option('display.float_format', '{:,.2f}'.format) 
df=pd.read_csv('/datasets/real_estate_data.csv', sep='\t')
df.info()
```

```python
df.head()
```

По информации info() и имеющейся документации можно первично оценить данные в столбцах и наметить фронт работ для предобработки данных. Всего в датасете 23 699 строк. И в ключевых для анализа, таких как total_area, last_price, first_day_exposition, floor, rooms - нет пропусков. 

first_day_exposition - поменям на формат даты.

В ceiling_height больше половины пропусков. Высота потолков - это один из параметров, которые нас просят проанализировать и оценить его влияние на цену. Наиболее подходящим было бы заполнение модой. Но мы его не будем заполнять для целей данного исследования, только избавимся от аномалий, чтобы не исказить его влияние на цену квартиры.

Floors_total - мало пропусков, ни на что не повлияют. Эти данные нам будут необходимы для разделения этажей на первый, последний, другой. Там где не указано количество этажей, поставим пометку 'не указан'.

living_area и kitchen_area заполним пропуски с помощью медианного коэффициента к общей площади с группировкой по количеству квартир.

is_apartment - поставим False вместо пропусков

locality_name - важный столбец для исследования, строки с немногочисленными пропусками удалим.

airports_nearest - это картографические данные, пропусков много. Не участвуют в дальнейшем анализе, оставим.

cityCenters_nearest - много пропусков. Пропуски оставим, чтобы не повлияли на анализ влияние этого параметра на цену.

parks_around3000 и ponds_around3000 - пропуски заменим на 0 (в радиусе 3км - нет парков и водоемов)

parks_nearest и ponds_nearest - оставим незаполненными, невозможно достоверно заполнить.

days_exposition - пропусков немного. Возможно означают, что на дату выгрузки сделки не были совершены. Заполним расчетным показателем first_day_exposition - дата выгрузки (последняя дата).

Для числовых данных в столбцах без пропусков (кроме метража) тип поменяем на целочисленный. Проанализируем аномалии.

Предварительно можно утверждать, что данных достаточно для исследования. Но перед началом анализа, нужно провести предобрабоку этих данных: заполнить пропуски, удалить дубликаты, поменять тип, выявить аномалии.

## Предобработка данных

### Locality_name

```python
round(df['locality_name'].isna().sum()/len(df), 3) #доля пропусков,%
```

```python
df.dropna(subset=['locality_name'], inplace=True) # удаляем строки с пропусками в locality_name
print('Количество строк после:', df.shape[0]) #проверка
```

Теперь посмотрим пропуски в каких столбцах не критичны для дальнейшего исследования и их можно заменить. Пропуски в количест ве балконов заменяем нулями. Если пропущено расстояние до аэропорта, заменяем медианным для данного региона. Если есть пропуски в количестве водоемов и парков в радиусе 3км, тоже заменяем 0. Пропуски в расстоянии до ближайшего водоема/парка оставляем - считая,что если не указано, значит их нет по близости. Пропуски в других столбцах оставляем, так они будут важны для исследовательского анализа данных и поэтому будут рассмотрены в следующем разделе. В столбце is_apartment, пропущенное значение принимаем за False.

### Balcony

```python
df['balcony']=df['balcony'].fillna(0) 
```

### Parks, ponds

```python
df['parks_around3000']=df['parks_around3000'].fillna(0) 
df['ponds_around3000']=df['ponds_around3000'].fillna(0) 
```

### Locality_name

```python
df['locality_name'].unique() #вывлдим уникальные значения locality_name
```

```python
df['locality_name'].nunique() #их количество
```

```python
#уберем типы наседенных пунктов из названий

for name in ['поселок ', 'деревня ', 'городской ', 'посёлок ', 'городского типа ', 
             'имени ', 'коттеджный ', 'садовое', 'товарищество ', 'садоводческое ', 
             'некоммерческое ', 'станции ', 'железнодорожной ', 'село ']:
         df['locality_name'] = df['locality_name'].str.replace(name, '')
```

```python
df['locality_name'].nunique() #количество населенных пунктов после
```

```python
df['locality_name'].value_counts().tail(10) #много населенных пунктов встерчаются по одному разу
```

Проанализируем 'locality_name'. Удалив типы населенных пунктов из названий мы сократили список за счет удаления неявных дубликатов с 364 до 309. Также видно, что много населенных пунктов встречаются один раз, поэтому не получится избавиться от пропусков в расстоянии до ближайшего парка или водоема, заменив их медианным расстоянием по населенному пункту. Данные пропуски оставляем.

### Is_apartment

```python
df['is_apartment']=df['is_apartment'].fillna(False) 
```

### Изменение типов данных

Где это на данном этапе возможно изменим тип данных на целочисленный.

```python
int_list = ['last_price','balcony', 'parks_around3000', 'ponds_around3000']
for item in int_list:
        df[item] = df[item].astype('int')
```

Переводим стодбец Дата публикации в формат Дата и время

```python
df['first_day_exposition']= pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d')
df['first_day_exposition'].head()
```

### Days_exposition

```python
# ищем последнюю дату публикации
(
    df
    .pivot_table(index='first_day_exposition', values='days_exposition', aggfunc='count')
    .sort_values('first_day_exposition', ascending=False)
)
```

```python
import datetime
```

```python
df['days_exposition']=df['days_exposition'].fillna((pd.to_datetime('2019-05-03')-df['first_day_exposition']).dt.days) 
```

Находим, что дата выгрузки (последняя в first_day_exposition) - 2019-05-03. Заменяем пропуски на разницу между этой датой и first_day_exposition. 

### Rooms

Проанализируем данные в столбце rooms. Странным кажется значение 0 в 197 строках. Это немного, на первый взгляд в основном кажется что небольшие квартиры, студии/со свободной планировкой. Поэтому заменим значение 0 на 1 для целей дальнейшего анализа. 

```python
df['rooms'].value_counts()
```

### Living_area, kitchen_area

Для того чтобы определить, когда можно считать, что продажи прошли очень быстро, а когда необычно долго, посторим диаграмму размаха и посчитаем гранцицы выбросов.  Если
правый «ус» длиннее максимума, то он заканчивается максимумом. То же —
для минимума и левого уса.

```python
q1 = df['days_exposition'].quantile(0.25) #первый квартиль
q3 = df['days_exposition'].quantile(0.75) #третий квартиль
iqr = q3 - q1 #межквартильный размах
down_border = q1 - iqr*1.5 #нижняя граница выбросов
up_border = q3 + iqr*1.5 #верхняя граница выбросов
if down_border<df['days_exposition'].min():
    print('Нижняя граница выбросов:', df['days_exposition'].min())
else: 
    print('Нижняя граница выбросов:', down_border)
    
if up_border>df['days_exposition'].max():
    print('Верхняя граница выбросов:', df['days_exposition'].max())
else: 
    print('Верхняя граница выбросов:', up_border)
```

Учитывая, что в начале интенсивность продаж высока, то, будем считать, что нормальными считаются продажи с длительностью 1 день. Хотя, понятно, что они прошли очень быстро. Но это не говорит о недостоверности данных. А долгие продажи больше 528 дней, будем считать выбросами и удалим. По гистограмме видно, что где-то после 250 дней частота продаж замирает. ### Определение сегмента квартиры в центре

Для начала выясним, какая область входит в центр. Создадим столбец с расстоянием до центра в километрах: округлим до целых значений. После этого посчитаем среднюю цену для каждого километра. Построим график, который покажет, как цена зависит от удалённости от центра. Определите границу, где график сильно меняется, — это и будет центральная зона.

```python
#выберем квартиры в Спб
pete_df=good_df.query('locality_name=="Санкт-Петербург"')
```

```python
#отключаем предупреждения
import warnings
warnings.filterwarnings('ignore')
```

```python
#выберем предложения, где указано расстояние до центра
pete_df.dropna(subset=['cityCenters_nearest'], inplace=True)
```

```python
#создадим столбец с расстоянием в километрах
pete_df['center_km']=(pete_df['cityCenters_nearest']/1000).round().astype(int)
```

```python
pete_df.head()
```

```python
#построим график зависимости цены от удаленности от центра
(pete_df
    .pivot_table(index='center_km', values='price_m')
    .plot(grid=True, kind='area', ylim={70000,120000}, xticks=pete_df['center_km'].unique())
)
```

```python
def rare_values(df, item):
    q1 = df[item].quantile(0.25) #первый квартиль
    q3 = df[item].quantile(0.75) #третий квартиль
    iqr = q3 - q1 #межквартильный размах
    down_border = q1 - iqr*1.5 #нижняя граница выбросов
    up_border = q3 + iqr*1.5 #верхняя граница выбросов
    return down_border, up_border
    
rare_values(pete_df, 'center_km')
```

**Вывод.** Есть необъяснимое падение на 3 км от центра. Принимая во внимание, что нижняя граница выбросов, это 2.5 км, данных в этом радиусе мало для достоверных выводов. После 8го километра наблюдается резкое падение. Этот радиус и будем считать центральным районом города.
